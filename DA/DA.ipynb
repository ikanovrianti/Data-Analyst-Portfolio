# Import Dataset 

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import pandasql as ps 
import matplotlib.pyplot as plt
%matplotlib inline
```

# Mempersiapkan Data
### Import Dataset

```python
data = pd.read_csv('DataAnalyst.csv')
data.head()
```

```python
data.info()
```

### Hapus Column yang tidak penting

```python
data = data.drop(["Unnamed: 0", "Competitors", "Easy Apply"], axis =1)
```

```python
data.head()
```

### Mencari Data NULL

```python
print(data.isnull().any()) #jika terdapat data berisi NULL maka hasilnya ada yg "True"
```

### Menghapus Data NULL

```python
data = data.dropna()
data.where(data == '-1')
data
```

### Terdapat beberapa Nilai "-1" pada Data. Untuk memperbaikinya, ganti nilai "-1" dengan "NA" 

```python
data = data.replace(to_replace = [-1, "-1"], value = "NA")
data.columns = data.columns.str.replace(' ', '_')
data = data[data.Sector != "NA"] #exclude all NA rows within "Sector" column
data.head()
```

```python
salary_estimate_adjustment = data['Salary_Estimate'].str.split('-', expand = True) #split numeric value
data['MinSalary'] = pd.to_numeric(salary_estimate_adjustment[0].str.extract('(\d+)', expand = False)) #buat column MinSalary berdasarkan kolom Salary_Estimate
data['MaxSalary'] = pd.to_numeric(salary_estimate_adjustment[0].str.extract('(\d+)', expand = False))
data.head()
```

### Membuat Visualisasi Data

#### 1. Mencari Top 20 Bidang Pekerjaan yang paling banyak di Data Analyst

```python
data['Job_Title'].value_counts()[:20]
```

```python
plt.figure(figsize=(15,7.5))

color_blind_palette = ["#DC267F", "#785EF0", "#648FFF", "#FE6100", "#FFB000", "#E80E8D", "#2E8766", "#95ABAC", "#65B925", "#906A42", "#C4D537", "#344A52", "#6C6E06", "#1C4DD2", "#216E00", "#2E03E5", "#A94424", "#7F6EA9", "#9B8453", "#380721"]

da_terbanyak = sns.barplot(x=data['Job_Title'].value_counts().head(20).index,
                                      y=data['Job_Title'].value_counts().head(20).values, palette=color_blind_palette)

plt.xticks(rotation='vertical')
plt.xlabel('Job_Title', fontsize=12.5)
plt.ylabel('Count', rotation=0, ha='right', fontsize=12.5)
plt.figtext(.5,.9,"Top 20 Bidang Pekerjaan yang paling banyak di DA", fontsize=20, fontweight='bold', fontname='helvetica', ha='center') #formatted title

for patch in da_terbanyak.patches:
             da_terbanyak.annotate("%.0f" % patch.get_height(), (patch.get_x() + patch.get_width() / 2., patch.get_height()),
                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                 textcoords='offset points')
plt.show()
```

### 2. Mencari TOP 20 Pekerjaan dengan gaji tertinggi pada perusahaan dengan peringkat teratas

```python
df_highest_salary_top_comp = """SELECT Job_Title, Company_Name, Sector,
                                    MAX(MaxSalary), MAX(Rating)
                                    FROM data
                                    GROUP BY Job_Title
                                    ORDER BY MaxSalary DESC, Rating DESC
                                    LIMIT 20"""
ps.sqldf(df_highest_salary_top_comp)
```

### 3. Sektor yang paling membutuhkan banyak Data Analyst

```python
df_sektor_banyak_kebutuhan = """SELECT Sector, COUNT(*) AS Count
                            FROM data
                            GROUP BY Sector
                            ORDER BY count DESC
                            LIMIT 20"""
ps.sqldf(df_sektor_banyak_kebutuhan)
```

```python
plt.figure(figsize=(15,7.5)) 
plt.xticks(rotation='vertical') 

color_blind_palette = ["#DC267F", "#785EF0", "#648FFF", "#FE6100", "#FFB000", "#E80E8D", "#2E8766", "#95ABAC", "#65B925", "#906A42", "#C4D537", "#344A52", "#6C6E06", "#1C4DD2", "#216E00", "#2E03E5", "#A94424", "#7F6EA9", "#9B8453", "#380721"]

plot_sektor_kebutuhan = sns.barplot(x=data['Sector'].value_counts().head(20).index,
                                  y=data['Sector'].value_counts().head(20).values, palette=color_blind_palette)

plt.xlabel('Sector', fontsize=12.5)
plt.ylabel('Count', rotation=0, ha='right', fontsize=12.5)
plt.figtext(.5,.9,"Sektor dengan Kebutuhan DA Terbanyak", fontsize=20, fontweight='bold', fontname='helvetica', ha='center') #formatted title

for patch in plot_sektor_kebutuhan.patches:
    plot_sektor_kebutuhan.annotate("%.0f" % patch.get_height(), (patch.get_x() + patch.get_width() / 2., patch.get_height()),
                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                 textcoords='offset points')

plt.show()
```

### 4. Daftar Kota dengan pekerjaan Data Analyst yang paling banyak

```python
df_city = """SELECT Location, COUNT(*) AS Job_Count
                FROM data
                GROUP BY Location
                ORDER BY Job_Count DESC
                LIMIT 20"""
ps.sqldf(df_city)
```

```python
plt.figure(figsize=(15,7.5))
plt.xticks(rotation='vertical')

color_blind_palette = ["#DC267F", "#785EF0", "#648FFF", "#FE6100", "#FFB000", "#E80E8D", "#2E8766", "#95ABAC", "#65B925", "#906A42", "#C4D537", "#344A52", "#6C6E06", "#1C4DD2", "#216E00", "#2E03E5", "#A94424", "#7F6EA9", "#9B8453", "#380721"]

plot_city = sns.barplot(x=data['Location'].value_counts().head(20).index,
                        y=data['Location'].value_counts().head(20).values, palette=color_blind_palette)
plt.xlabel('Location', fontsize=12.5)
plt.ylabel('Count', rotation=0, ha='right', fontsize=12.5)
plt.figtext(.5,.9,'Kota dengan DA Terbanyak', fontsize=20, fontweight='bold', fontname='helvetica', ha='center') #formatted title

for patch in plot_city.patches:
             plot_city.annotate("%.0f" % patch.get_height(), (patch.get_x() + patch.get_width() / 2., patch.get_height()),
                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                 textcoords='offset points')

plt.show()
```

### 5. Mencari korelasi antara peringkat/rating perusahaan dengan gaji yang ditawarkan

```python
corr_data = data
corr_data["AvgSalary"] = corr_data[['MinSalary', 'MaxSalary']].mean(axis=1) #membuat kolom avgsalary utk menghitung rata2 dari MinSalary & MaxSalary
corr_data["Rating"] = pd.to_numeric(corr_data.Rating, errors='coerce') #convert typedata int kolom rating ke typedata numeric
corr_data.corr()
```

```python
plt.figure(figsize=(20,15))
plt.figtext(.5, .9, "Salary and Rating", fontsize=20, fontweight='bold', fontname='Helvetica', ha='right') #formatted title
sns.heatmap(corr_data.corr(), annot=True,cmap="PiYG")
```

Dari plotted headmap, terlihat bahwa hampir tidak ada korelasi (nilainya sangat dekat dengan 0) antara gaji min/maks atau rata-rata dan peringkat.
